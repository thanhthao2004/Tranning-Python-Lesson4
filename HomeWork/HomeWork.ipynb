# Bài tập Thực hành Trực quan hóa Dữ liệu

## Dữ liệu 'Data_sale.csv'
`Bộ dữ liệu bao gồm thông tin về doanh số bán hàng, số lượng khách hàng, sản phẩm, và vùng địa lý  `
- Date: Ngày cuối mỗi tháng từ tháng 1 đến tháng 10 năm 2023.
- Region: Vùng địa lý (Bắc, Nam, Đông, Tây).
- Product: Loại sản phẩm (A, B, C, D).
- Sales: Doanh số bán hàng (giá trị định lượng).
- Customers: Số lượng khách hàng.
- Profit: Lợi nhuận thu được từ doanh số.

## Bài tập 1: Biểu đồ Scatter Plot
Hãy vẽ biểu đồ scatter plot thể hiện mối quan hệ giữa số lượng khách hàng và doanh số:

### Yêu cầu:
1. Trục x: Số lượng khách hàng (Customers)
2. Trục y: Doanh số (Sales)  
4. Thêm:
    - Tiêu đề biểu đồ
    - Nhãn trục x, y
    - Chú thích (legend) cho các vùng
5. Nhận xét đưa ra insights

## Bài tập 2: Biểu đồ Box Plot
Tạo box plot so sánh phân phối doanh số giữa các sản phẩm:

### Yêu cầu:
1. Trục x: Sản phẩm (Product)
2. Trục y: Doanh số (Sales)
3. Sử dụng màu sắc khác nhau cho mỗi sản phẩm
4. Thêm tiêu đề phù hợp cho biểu đồ
5. Nhận xét đưa ra insights

## Bài tập 3: Biểu đồ Tròn (Pie Chart)
Vẽ biểu đồ tròn thể hiện tỷ trọng doanh số của từng vùng:

### Yêu cầu:
1. Tính tổng doanh số theo vùng
2. Tạo pie chart với:
    - Các phần được tô màu khác nhau
    - Nhãn phần trăm cho mỗi phần
3. Thêm tiêu đề cho biểu đồ
4. Nhận xét đưa ra insights

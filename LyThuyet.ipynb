```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

`Bộ dữ liệu bao gồm thông tin về doanh số bán hàng, số lượng khách hàng, sản phẩm, và vùng địa lý  `
- Date: Ngày cuối mỗi tháng từ tháng 1 đến tháng 10 năm 2023.
- Region: Vùng địa lý (Bắc, Nam, Đông, Tây).
- Product: Loại sản phẩm (A, B, C, D).
- Sales: Doanh số bán hàng (giá trị định lượng).
- Customers: Số lượng khách hàng.
- Profit: Lợi nhuận thu được từ doanh số.

```python
# pandas có hỗ trợ đọc các loại file như excel hay csv
df = pd.read_csv('Data_sale.csv')
```

# Biểu đồ Đường (Line Chart) - Sales Trend Over Time

```python
# Vẽ biểu đồ Line Chart
plt.figure(figsize=(12, 6))  # Kích thước biểu đồ
plt.plot(df['Date'], df['Sales'], color='blue', linestyle='-', marker='o', markersize=5,
         linewidth=1.5, label="Sales Over Time")

# Tiêu đề và nhãn trục
plt.title("Sales Trend Over Time")
plt.xlabel("Date")
plt.ylabel("Sales")
plt.xticks(rotation=45)  # Xoay nhãn trục x để dễ đọc hơn
plt.grid(True)  # Hiển thị lưới
plt.legend()  # Hiển thị chú thích cho đường
plt.tight_layout()  # Đảm bảo mọi thứ không bị cắt khi hiển thị

# Hiển thị biểu đồ
plt.show()
```

- Xu hướng doanh số: Biểu đồ đường cho thấy doanh số có sự dao động theo thời gian, với một số đỉnh và đáy rõ ràng. Điều này cho thấy doanh số không ổn định, có thể do yếu tố mùa vụ hoặc các sự kiện ảnh hưởng đến nhu cầu.
- Các đỉnh cao doanh số: Có những khoảng thời gian doanh số tăng đột biến, điều này có thể trùng với các đợt khuyến mãi hoặc mùa bán hàng cao điểm. Công ty có thể tận dụng những thời điểm này để thúc đẩy doanh số qua các chiến dịch quảng cáo hoặc khuyến mãi mạnh mẽ hơn.
- Khoảng thời gian doanh số thấp: Một số đoạn thời gian doanh số giảm đáng kể. Điều này có thể là do yếu tố chu kỳ hoặc thời điểm thấp điểm. Để cải thiện doanh số trong các thời điểm này, công ty có thể cân nhắc các chiến lược giảm giá hoặc tung ra sản phẩm mới để kích cầu.

# Biểu đồ Cột (Bar Chart) - Sales Over Time

```python
# Vẽ biểu đồ cột (Bar Chart)
plt.figure(figsize=(20, 10))  # Kích thước biểu đồ lớn hơn để dễ đọc
plt.bar(df['Date'], df['Sales'],color='red')  # Vẽ cột biểu diễn doanh số theo ngày
plt.title("Sales Bar Chart Over Time")
plt.xlabel("Date")
plt.ylabel("Sales")
plt.xticks(rotation=45)  # Xoay nhãn trục x để dễ đọc hơn
plt.tight_layout()  # Đảm bảo bố cục gọn gàng, không bị cắt nhãn
plt.show()
```

=> Doanh số có xu hướng biến động khá nhiều theo thời gian. Các đỉnh cao trong doanh số cho thấy có thể có những thời điểm nhất định với nhu cầu cao hơn (có thể là do các đợt khuyến mãi hoặc mùa bán hàng).
- Một số ngày có doanh số đặc biệt thấp hơn, cho thấy nhu cầu giảm hoặc sự thiếu hụt trong cung cấp sản phẩm.

#  Biểu đồ Tròn (Pie Chart) - Sales Distribution by Product

```python
sales_by_product = df.groupby('Product')['Sales'].sum()

# Vẽ biểu đồ tròn đơn giản hơn (Pie Chart)
plt.pie(sales_by_product, labels=sales_by_product.index, autopct='%1.1f%%')

plt.title("Sales Distribution by Product")
plt.show()
```

Biểu đồ tròn thể hiện tỷ lệ đóng góp của từng sản phẩm vào tổng doanh số. Một số sản phẩm có doanh số đóng góp lớn hơn, có thể là các sản phẩm chủ lực của công ty.
Các sản phẩm có doanh số thấp hơn có thể là do thị trường hạn chế hoặc do mức độ phổ biến của sản phẩm không cao.


# Biểu đồ Phân tán (Scatter Plot) - Sales vs. Customers with Profit

```python

# Biểu đồ phân tán (Scatter Plot) cải tiến với Seaborn cho mối quan hệ giữa số lượng khách hàng và doanh số
plt.figure(figsize=(8, 6))
sns.scatterplot(data=df, x='Customers', y='Sales', hue='Region', style='Product', s=100)
plt.title("Enhanced Scatter Plot of Sales vs Customers")
plt.xlabel("Customers")
plt.ylabel("Sales")
plt.legend(title="Region/Product")
plt.show()
```

=> Biểu đồ phân tán cho thấy mối quan hệ giữa số lượng khách hàng và doanh số. Nhìn chung, có một xu hướng tích cực giữa số lượng khách hàng và doanh số, tức là khi số lượng khách hàng tăng, doanh số cũng có xu hướng tăng.
- Các điểm dữ liệu có màu sáng hơn (tương ứng với lợi nhuận cao) xuất hiện ở nhiều vị trí, cho thấy không phải lúc nào doanh số cao cũng mang lại lợi nhuận cao. Lợi nhuận có thể bị ảnh hưởng bởi nhiều yếu tố khác ngoài doanh số, chẳng hạn như chi phí sản xuất hoặc chính sách giá.

### Một chút nâng cao nhưng đa số scatter chart sẽ được sử dụng để biểu thị mối tương quan giữa 2 biến định lượng nhá ;3
#### Giờ chị sẽ randome 1 data. Hàm random có sẵn trong thư viên **Numpy** của Python

```python
#Cài đặt thư viện bằng cách gõ:
!pip install Numpy
```

```python
# Tạo dữ liệu mẫu
import numpy as np
np.random.seed(42)
x = np.random.rand(100) * 10  # Biến độc lập
y = 2.5 * x + np.random.randn(100) * 5  # Biến phụ thuộc (có thêm nhiễu)

# Vẽ scatter chart với hệ số tương quan
plt.figure(figsize=(8, 6))
sns.scatterplot(x=x, y=y, color='blue', label='Dữ liệu mẫu')
plt.title("Biểu đồ Scatter với Hệ số Tương quan", fontsize=14)
plt.xlabel("Biến X", fontsize=12)
plt.ylabel("Biến Y", fontsize=12)

# Tính hệ số tương quan
correlation_coefficient = np.corrcoef(x, y)[0, 1]
plt.text(1, 25, f"Hệ số tương quan: {correlation_coefficient:.2f}", fontsize=12, color='red')

plt.legend()
plt.grid(True, alpha=0.3)
plt.show()
```

# Biểu đồ Hộp (Box Plot) - Profit by Product

```python
# Biểu đồ hộp (Box Plot) cải tiến với Seaborn cho phân phối lợi nhuận theo sản phẩm
# Tạo thuộc tính hiển thị outliers
flierprops = dict(marker='o', color='red', alpha=0.6)

plt.figure(figsize=(8, 6))
sns.boxplot(
    data=df, 
    x='Product', 
    y='Profit', 
    palette="Set3", 
    showfliers=True,  # Hiển thị outliers
    flierprops=flierprops
)
plt.title("Enhanced Box Plot of Profit by Product with Outliers")
plt.xlabel("Product")
plt.ylabel("Profit")
plt.show()
```

=> Biểu đồ hộp cho thấy sự phân phối lợi nhuận khác nhau giữa các sản phẩm. Một số sản phẩm có lợi nhuận ổn định hơn (ít biến động trong hộp) trong khi các sản phẩm khác có lợi nhuận biến động lớn.
Có một số giá trị ngoại lệ, cho thấy một số thời điểm lợi nhuận từ các sản phẩm này tăng hoặc giảm đáng kể. Điều này có thể do nhu cầu thay đổi hoặc do các đợt khuyến mãi làm lợi nhuận dao động.
